rent/return vehicle system

```python
vehicles=[]

cars=[]

motorcycles=[]

heavy_vehicles=[]
```

```python
class Vehicle:
    def __init__(self, id, brand, model, year, color, daily_rental_price, max_days=14):
        self.id = id
        self.brand = brand
        self.model = model
        self.year = year
        self.color = color
        self.daily_rental_price = daily_rental_price
        self.max_days = max_days
        self.rented = False

    def add_to_list(self):
        """
        Add the object to the related list of vehicles.
        """
        vehicles.append(self)


    def is_rentable(self, requested_days):
        """
        Check if the number of days the user wants to rent the vehicle
        exceeds the maximum number of days the vehicle can be rented.
        """
        return requested_days <= self.max_days

    def calculate_total_cost(self, requested_days):
        """
        Calculate the total price based on the number of days entered by the user
        and the daily rental price of the vehicle.
        """
        total = self.daily_rental_price * requested_days
        print(f"You have to pay {total} for {requested_days} days")
        return total
    
    def change_rental_status(self, action):
      """
      Change the rented attribute of the instance based on the action parameter.
      """
      if action == "rent":
         if not self.rented:
            self.rented = True
            print(f"The vehicle model: {self.model} is now rented.")
         else:
            print(f"The vehicle model: {self.model} is already rented. You cannot rent this car now.")
      elif action == "return":
        if self.rented:
            self.rented = False
            print(f"The vehicle model: {self.model} is now returned and available to rent.")
        else:
            print(f"The vehicle model: {self.model} is not rented! It is already available.")
      else:
           print("Invalid action! Please use 'rent' or 'return'.")

    def display_details(self):
        """
        Display all details of the vehicle.
        """
        print(f"ID: {self.id}")
        print(f"Brand: {self.brand}")
        print(f"Model: {self.model}")
        print(f"Year: {self.year}")
        print(f"Color: {self.color}")
        print(f"Daily Rental Price: {self.daily_rental_price}")
        print(f"Max Days: {self.max_days}")
               
```

```python
class Car(Vehicle):
    def __init__(self, id, brand, model, year, color, daily_rental_price, max_days, num_doors, trunk_size, fuel_type):
        super().__init__(id, brand, model, year, color, daily_rental_price, max_days)
        self.num_doors = num_doors
        self.trunk_size = trunk_size  # size in liters
        self.fuel_type = fuel_type    # petrol, diesel, electric

    def add_to_list(self):
        """
        Add the car to the list of cars.
        """
        cars.append(self)

    def display_details(self):
        """
        Display all details of the car.
        """
        print(f"ID: {self.id}")
        print(f"Brand: {self.brand}")
        print(f"Model: {self.model}")
        print(f"Year: {self.year}")
        print(f"Color: {self.color}")
        print(f"Daily Rental Price: {self.daily_rental_price}")
        print(f"Max Days: {self.max_days}")
        print(f"Number of Doors: {self.num_doors}")
        print(f"Trunk Size: {self.trunk_size} liters")
        print(f"Fuel Type: {self.fuel_type}")
        print(f"Rented: {self.rented} \n")    
    
    
```

```python
class Motorcycle(Vehicle):
    def __init__(self, id, brand, model, year, color, daily_rental_price, max_days, has_sidecar, engine_capacity, type):
        super().__init__(id, brand, model, year, color, daily_rental_price, max_days)
        self.has_sidecar = has_sidecar
        self.engine_capacity = engine_capacity  # capacity in cc
        self.type = type  # e.g., sport, cruiser

    def add_to_list(self):
        """
        Add the motorcycle to the list of motorcycles.
        """
        motorcycles.append(self)

    def display_details(self):
        """
        Display all details of the motorcycle.
        """
        print(f"ID: {self.id}")
        print(f"Brand: {self.brand}")
        print(f"Model: {self.model}")
        print(f"Year: {self.year}")
        print(f"Color: {self.color}")
        print(f"Daily Rental Price: {self.daily_rental_price}")
        print(f"Max Days: {self.max_days}")
        print(f"Has Sidecar: {'Yes' if self.has_sidecar else 'No'}")
        print(f"Engine Capacity: {self.engine_capacity} cc")
        print(f"Type: {self.type}")
        print(f"Rented: {self.rented} \n")            
       
```

```python
class HeavyVehicle(Vehicle):
    def __init__(self, id, brand, model, year, color, daily_rental_price, max_days, max_load_capacity, number_of_axles, trailer_length):
        super().__init__(id, brand, model, year, color, daily_rental_price, max_days)
        self.max_load_capacity = max_load_capacity  # capacity in kg
        self.number_of_axles = number_of_axles
        self.trailer_length = trailer_length  # length in meters

    def add_to_list(self):
        """
        Add the heavy vehicle to the list of heav vehicles.
        """
        heavy_vehicles.append(self)
    def display_details(self):
        """
        Display all details of the heavy vehicle.
        """
        print(f"ID: {self.id}")
        print(f"Brand: {self.brand}")
        print(f"Model: {self.model}")
        print(f"Year: {self.year}")
        print(f"Color: {self.color}")
        print(f"Daily Rental Price: {self.daily_rental_price}")
        print(f"Max Days: {self.max_days}")
        print(f"Max Load Capacity: {self.max_load_capacity} kg")
        print(f"Number of Axles: {self.number_of_axles}")
        print(f"Trailer Length: {self.trailer_length} meters")
        print(f"Rented: {self.rented} \n")
    
        
 
```

```python
def display_available_vehicles(vehicles):
    """
    Display details of available vehicles for rent.
    """
    for vehicle in vehicles:
        if not vehicle.rented:
            vehicle.display_details()

def display_rented_vehicles(vehicles):
    """
    Display details of rented vehicles for return.
    """
    for vehicle in vehicles:
        if vehicle.rented:
            vehicle.display_details()
```

```python
def handle_rental(vehicle_list, vehicle_type):
    '''
    check if user can rent a vehicle according to requested days and then calculate the total cost for rent
    '''

    display_available_vehicles(vehicle_list)

    id_input = input(f"Please enter the {vehicle_type} Id that you want to rent: ").strip()
    requested_days = input("How many days do you need the vehicle: ").strip()
    
    for vehicle in vehicle_list:
        if vehicle.id == int(id_input): 
            if vehicle.is_rentable(int(requested_days)):
               vehicle.change_rental_status("rent")
               vehicle.calculate_total_cost(int(requested_days))
               return
            else:
                 print(f"can not rent this vehicle for more than {vehicle.max_days} days")
                 return
        

def handle_return(vehicle_list, vehicle_type):

    display_rented_vehicles(vehicle_list)
    
    id_input = input(f"Please enter the {vehicle_type} Id that you want to return: ").strip()
    
    for vehicle in vehicle_list:
        if vehicle.id == int(id_input):
            vehicle.change_rental_status("return")
            return        
```

```python
action = input("What do you want to do? (rent/return): ").lower().strip()

if action not in ["rent", "return"]:
    print("Invalid action! Please enter 'rent' or 'return'.")

else:
    vehicle_type = input('''What type of vehicle do you want to rent/return?
                       1 - Car
                       2 - Motorcycle
                       3 - Heavy Vehicle
                       Enter 1, 2, or 3: ''').strip()

    if vehicle_type not in ["1", "2", "3"]:
       print("Invalid input! Please enter 1, 2, or 3.")
    else:
        if vehicle_type == "1": # action for car
            if action == "rent":
               handle_rental(cars, "Car")
            else:
               handle_return(cars, "Car")
        
        elif vehicle_type == "2": # action for motorcycle
            if action == "rent":
               handle_rental(motorcycles, "Motorcycle")
            else:
               handle_return(motorcycles, "Motorcycle")

        else:  # action for heavyvehicle
            if action == "rent":
               handle_rental(heavy_vehicles, "Heavevehicle")
            else:
               handle_return(heavy_vehicles, "Heavyvehicle")
         
        


     

```
